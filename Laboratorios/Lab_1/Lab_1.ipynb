## Parte 1: Conocimiento Teórico (2 pts)

Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?

2. Nombre y explique 3 diferencias entre `Git` y `Github`.

3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.

4. ¿Que elementos componen a la estructura de datos de `Git`?

5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?

## **Respuestas:**

1.   `Git` es un sistema de control de versiones de código abierto que permite rastrear y gestionar cambios en el código fuente de un proyecto. `Git` almacena los cambios en un repositorio, que se puede mantener localmente o en un servidor remoto. Los usuarios pueden trabajar en diferentes ramas o versiones del código y combinarlas.
2.   - La principal diferencia es que Git es un sistema de control de versiones distribuido, en cambio, Github es un proovedor de este servicio.
    - Github requiere que se registre una cuenta para utilizarlo, mientras que eso no es necesario con Git.
    - Github puede ser utilizado remotamente debido a que esta alojado en la nube, en cambio git solo permite clonar repositorios desde servidores remotos como Github.
3.   Un `hash` es una cadena de caracteres que se utiliza para identificar de manera única un `commit` en un repositorio. El `hash` se calcula a partir de la información del `commit`, y se representa como una cadena hexadecimal de 40 caracteres.
4.   La estructura de datos de `Git` está compuesta por un repositorio central que contiene commits, ramas, etiquetas y un árbol de directorios. 
    - Cada commit es un conjunto de cambios realizados en un archivo o conjunto de archivos. 
    - Las ramas permiten trabajar en diferentes versiones del proyecto sin afectar la versión principal. 
    - Las etiquetas se utilizan para marcar versiones específicas del proyecto. 
    - El árbol de directorios representa la organización de los archivos y directorios en el proyecto.
5.   Para crear un repositorio en `Git`, se utiliza el comando `git init`, que inicializa un nuevo repositorio en el directorio actual. Para hacer seguimiento a cambios en los archivos, se utiliza el comando `git add` para agregar archivos al área de preparación, y luego el comando `git commit` para confirmar los cambios y crear un nuevo commit en el historial del repositorio.

Responda si las siguientes afirmaciones son Verdaderas (V) o Falsas (F). Justifique en **cualquiera** de los casos (0.2 pts c/u):


6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.
7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.
8. Cada miembro del equipo trabaja en la misma copia del código fuente usando `Git`.
9. En `Git` es posible deshacer cambios y volver a versiones anteriores del código.
10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.

## **Respuestas:**

6.   **Falso**. `Git` no siempre resuelve automáticamente los problemas de línea al hacer merge entre diferentes ramas. En algunos casos, pueden surgir conflictos que requieren intervención manual para solucionar los problemas.
7.   **Falso**. `git clone` crea una copia local completa de un repositorio remoto, mientras que `git pull` actualiza una copia local existente con los cambios más recientes del repositorio remoto.
8.   **Falso**. Cada usuario trabaja en su propia copia del código fuente, para sincronizar el repositorio remoto se pueden utilizar comandos como `git push` y `git pull`. 
9.   **Verdadero**. `Git` permite deshacer cambios y volver a versiones anteriores del código a través de comandos como `git reset` y `git revert`.
10.   **Falso**. `git commit` se utiliza para guardar los cambios en el repositorio local, pero no se envían automáticamente al repositorio remoto, para eso se utiliza el comando `git push`.

7. Cree una rama nueva y proponga una solución al problema anterior, adjuntando el código corregido y señalando las líneas modificadas. (0.5 pts)

Se creo una nueva rama utilizando el codigo <code>git checkout -b fix-conflict</code>, el cual ademas se encarga de cambiar automaticamente a esta. Para poder solucionar el problema anterior se debe corregir el codigo incorrecto en el la rama fix-ifs, donde no se realizo la separacion del modulo y no existe el modulo animales.py,  de modo que el codigo del archivo main.py quedara de la siguiente manera:



```
def animales(input):
    
    '''
    Como hace el animalito?
    '''
    
    if input.lower() == 'gato':
        print('miau')
    elif input.lower() == 'perro':
        print('guau')
    elif input.lower() == 'fox':
        print('https://www.youtube.com/watch?v=jofNR_WkoCE')
    else:
        raise ValueError('animal no reconocido :(')

if __name__ == '__main__':
    print("Ingrese el animalito para saber que dice:")
    animal = input()
    animales(animal)
```
De este modo al realizar el merger no existiria un conflicto en el codigo y ademas este podria correrse de manera exitosa pues contiene las modificaciones necesarias para esto.
