### 1.3.1 Transformaci칩n de notas a "Nota Chilena" [0.5 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci칩n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci칩n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci칩n entregada, cree una funci칩n que le permita pasar a la escala Chilena las notas que se escapan del rango [1,7] y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [X] Tranforme los valores a la escala chilena.

````python
df_grades.describe()
````

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
  dataframe = dataframe_in.copy()
  dataframe[columna] = (dataframe[columna] * 0.06 + 1).round(2)
  return dataframe
  
````

````python
df_grades = nota_chilena(df_grades, ['math score', 'reading score', 'writing score'])
df_grades
````

### 1.3.3 Reprobados y Destacados [1.5 Punto, 0.5 Cada DataFrame]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o y los que han tenido desempe침o de destacado. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar치 destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci칩n que permita obtener tres Dataframes: 
- El primero deber치 poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser치 un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [X] Obtener DataFrame con los alumnos que reprobaron el a침o.
- [X] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [X] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):

    dataframe = dataframe_in.copy()

    reprobados = dataframe.loc[dataframe['GPA'] < 4].reset_index(drop=True)
    
    reprobados_R = dataframe.copy()
    reprobados_R.loc[reprobados_R['GPA'] < 4, 'GPA'] = 'R'  
    reprobados_R = reprobados_R.reset_index(drop=True)
    
    cols_notas = ['math score', 'reading score', 'writing score']
    destacados = dataframe[dataframe[cols_notas].gt(5.8).all(axis=1)].sort_values(by='GPA', ascending=False).reset_index(drop=True)
    return reprobados, reprobados_R, destacados
````

### 1.3.5 Quintiles de Rendimiento [0.5 Puntos]

El rector, intrigado por saber si existe una relaci칩n entre el nivel educacional de los padres y las notas de sus hijos les plantea estudiar dicha relaci칩n. Para esto, les solicita que desde el `DataFrame` que posee a todos los alumnos (`DataFrame` obtenido tras la limpieza), generen una nueva columna que clasifique en quintiles de rendimiento para sus alumnos: `['Baj칤simo', 'Bajo', 'Medio', 'Alto', 'Muy alto']`. (investigar `qcut()`)

Hecho esto, grafique a trav칠s de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Realizado el grafico, responda la siguiente pregunta: **쮼s posible visualizar alguna tendencia en el gr치fico?.**



**Respuesta:**

````python
# C칩digo para quintiles 
# La variable que contenga los quintiles debe llamarse GPA cuantiles para que el gr치fico funcione.
labels = ['Baj칤simo', 'Bajo', 'Medio', 'Alto', 'Muy alto']
df_grades['GPA cuantiles'] = pd.qcut(df_grades['GPA'], q=5, labels=labels)

# C칩digo de gr치fico 
fig = px.scatter(
    df_grades.sort_values(by="GPA"),
    x="GPA",
    y="parental level of education",
    hover_data=["gender"],
    color="GPA cuantiles",
    title="GPA por Nivel de educaci칩n de los padres",
    color_discrete_sequence=["tomato", "orange", "yellow", "greenyellow", "green"]
)

fig
````


Se puede ver que el m칤nimo GPA de los estudiantes, asciende de manera casi lienal a medida que aumenta el nivel educacional de los padres. Tambi칠n se puede ver una tendencia a estar m치s concentrados en GPA m치s altos los alumnos que tienen padres con mayor nivel eduacional. Y al rev칠s, los hijos que provienen de padres con bajo nivel eduacional tienen una dispersi칩n mayor en los datos. Esta dispersi칩n como se dijo antes, disminuye a medida que aumenta el nivel educacional de los padres. En resumen, si los padres del hijo vienen de un nivel educacional alto, es menos probable que tenga GPA baj칤simo.
