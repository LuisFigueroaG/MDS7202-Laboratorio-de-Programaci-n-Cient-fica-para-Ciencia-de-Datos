### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [X] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [X] Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`  seg칰n la siguiente conversi칩n: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [X] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [X] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [X] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como gu칤a para desarrollar este punto.

````python
# Calculamos el GPA
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].apply(lambda x: x.mean(), axis=1)
````

````python
# Diccionario para aplicar el mapeo
edu_map = {'some high school': 'school',
           'some college': 'school',
           'high school': 'school',
           "bachelor's degree": 'college',
           "associate's degree": 'college',
           "master's degree": 'postgraduate'}

# Mapeo
df_grades['simple parental level of education'] = df_grades['parental level of education'].map(edu_map)
````

````python
# Agrupar por race/ethnicity y simple parental level of education y obtener el promedio de las notas
mean_grades_df = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).mean()[['math score', 'reading score', 'writing score', 'GPA']]

# Redondeamos los promedios
mean_grades_df[['math score', 'reading score', 'writing score', 'GPA']] = mean_grades_df[['math score', 'reading score', 'writing score', 'GPA']].round(2)
````

````python
# Agrupar por race/ethnicity y simple parental level of education y obtener el count
count_df = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')
count_df = count_df.set_index(['race/ethnicity', 'simple parental level of education'])
````

````python
# Concatenamos los dataframes
result_df = pd.concat([mean_grades_df, count_df], axis=1)
````

````python
# Porcentaje de alumnos respecto al total (como string)
total_alumnos = df_grades.shape[0]
result_df['percentage'] = (result_df['count'] / total_alumnos * 100).apply(lambda x: '{:.2f}%'.format(x))
````

````python
# Se carga el dataframe para las otras notas
df_other =  pd.read_csv('./other_grades.csv')
df_other
````

````python
outer_join = pd.merge(df_grades, df_other, on='names', how='outer')
left_join = pd.merge(df_grades, df_other, on='names', how='left')
right_join = pd.merge(df_grades, df_other, on='names', how='right')
inner_join = pd.merge(df_grades, df_other, on='names', how='inner')
````

````python

print('OUTER JOIN\n')
print('Cantidad de filas:', outer_join.shape[0])
print('Cantidad de valores nulos:', outer_join.isnull().sum())
display(outer_join.head(10))

print('LEFT JOIN\n')
print('Cantidad de filas:', left_join.shape[0])
print('Cantidad de valores nulos:', left_join.isnull().sum())
display(left_join.head(10))

print('RIGHT JOIN\n')
print('Cantidad de filas:', right_join.shape[0])
print('Cantidad de valores nulos:', right_join.isnull().sum())
display(right_join.head(10))

print('INNER JOIN\n')
print('Cantidad de filas:', inner_join.shape[0])
print('Cantidad de valores nulos:', inner_join.isnull().sum())
display(inner_join.head(10))
````

> **Justificaci칩n:**


Al observar los dataframes resultantes de los distintos tipos de join, podemos observar que todos los estudiantes del dataframe df_grades tienen registros para sus puntajes de historia y ciencias, sin embargo, en el caso contrario no es asi, es decir, hay estudiantes de df_others que no tienen registros en df_grades, es decir no hay registros de sus notas en otras materias o de su raza o registros de sus padres. Teniendo en cuenta lo anterior, la decision de que join utilizar dependara simplemente del objetivo del join. En este caso se asumira que para poder realizar un analisis exitoso es necesario tener registros de la raza del estudiantes, del nivel de estudios de sus padres y de los puntajes en todas las pruebas. Por lo tanto, se quiere obtener un dataframe donde no se obtengan valores nulos en ninguna columna. Al observar las composiciones de los dataframes, podemos ver que se obtienen estos resultados realizando un left join y un inner join. Sin embargo, el left join funciona solo porque todos los estudiantes del dataframe df_grades tienen registros para sus pruebas de historia y ciencias en el dataframe df_other. En el caso de que esto no fuera asi se verian registros donde algunos estudiantes tendrian valores nulos en sus pruebas de historia y ciencias. Debido a esto, la mejor decision seria utilizar un inner join, donde solo se mantendran los registros de estudiantes que tengan informacion en ambos dataframes.

````python

fig = sp.make_subplots(rows=2, cols=2, subplot_titles=("Math Score", "Writing Score", "Reading Score", "GPA"))

for idx, score_var in enumerate(['math score', 'writing score', 'reading score', 'GPA']):
    data = []
    group_labels = []
    for course in df_grades['test preparation course'].unique():
        temp_df = df_grades[df_grades['test preparation course'] == course]
        data.append(temp_df[score_var])
        group_labels.append(course)

    sub_fig = ff.create_distplot(data, group_labels, show_hist=False, show_rug=False)

    
    for trace_idx, trace in enumerate(sub_fig['data']):
        trace.showlegend = idx == 0 
        fig.add_trace(trace, row=1 if idx < 2 else 2, col=idx % 2 + 1)

fig.update_layout(title="Distribuci칩n de puntajes por curso de preparaci칩n", showlegend=True, width=800, height=800)

fig.show()
````
